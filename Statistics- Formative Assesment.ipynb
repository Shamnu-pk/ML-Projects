# IQR 

```python
q1=df1.price_per_sqft.quantile(0.25)
print("q1 =",q1)
q3=df1.price_per_sqft.quantile(0.75)
print("q1 =",q3)
IQR=q3-q1
print("IQR =",IQR)
```

```python
df1['price_per_sqft'].describe()
```

```python
lower_whisker=q1-1.5*IQR
upper_whisker=q3+1.5*IQR
lower_whisker,upper_whisker
```

```python
remov_out=df1[(df1.price_per_sqft<lower_whisker) | (df1.price_per_sqft>upper_whisker)]
```

```python
remov_out.skew(numeric_only=True)
```

```python
def remove_outliers(df, numerical):
    filterd=df.copy()
    
    for col in numerical:
        q1=df[col].quantile(0.25)
        q3=df[col].quantile(0.75)
        IQR=q3-q1
        
        lower_whisker=q1-1.5*IQR
        upper_whisker=q3+1.5*IQR
        
        filterd=filterd[(filterd[col]>= lower_whisker) & (filterd[col]<= upper_whisker)]
    return filterd
```

```python
dff=remove_outliers(df,['total_sqft', 'bath', 'price', 'bhk','price_per_sqft'])
dff.skew(numeric_only=True)
```

```python
q1=dff.price.quantile(0.25)
q3=dff.price.quantile(0.75)
IQR=q3-q1

lower_whisker=q1-1.5*IQR
upper_whisker=q3+1.5*IQR
lower_whisker,upper_whisker

remov_out=dff[(dff.price<lower_whisker) | (dff.price>upper_whisker)]
remov_out.skew(numeric_only=True)
```

```python
sns.boxplot(remov_out)
plt.show()
```

```python
sns.boxplot(remov_out["price"])
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(remov_out['price_per_sqft'], kde=True)
plt.title('Histogram of Price per Sqft')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
numeric_columns = remov_out.select_dtypes(include=['number'])

correlation_matrix = numeric_columns.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Heatmap of Numerical Columns')
plt.show()
```
